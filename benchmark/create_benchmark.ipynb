```python
import json
import hashlib

class JsonTemplate:
    def __init__(self):
        self.paste = {"origin_doc_name": None,
                      "ideal_context": None,
                      "question": None,
                      "answer": None}
        
    def paste_doc_name(self, name):
        self.paste['origin_doc_name'] = name.encode('utf-8').decode('utf-8')
        
    def paste_question(self, question):
        self.paste['question'] = question.encode('utf-8').decode('utf-8')

        
    def paste_answer(self, answer):
        self.paste['answer'] = answer.encode('utf-8').decode('utf-8')
        
    def paste_ideal_context(self, context):
        self.paste['ideal_context'] = context.encode('utf-8').decode('utf-8')
        
    def save(self):
        if None in self.paste.values():
            raise ValueError("Not all values were pasted")
        
        name = hashlib.sha1(str(self.paste).encode()).hexdigest()
        with open(f'QnA/{name}.json', 'w', encoding='utf-8') as file:
            json.dump(self.paste, file, indent=4, ensure_ascii=False)
        
        self.paste = {"origin_doc_name": None,
                      "ideal_context": None,
                      "question": None,
                      "answer": None}
t = JsonTemplate()
```
